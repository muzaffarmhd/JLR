```python
import jax.numpy as jnp
from jax import grad

def u(t, a):
    x = a * t**2
    y = 2*a*t
    return y**2 - 4 * a * x
du_dt = grad(u, argnums=0)

result = du_dt(1.0,1.0)

print(f"du/dt at t=1 and a=1 is {result}")
```

```python
def z(t):
    x = jnp.cos(t)
    y = jnp.sin(t)
    return x**2 + y**2
dz_dt = grad(z, argnums=0)

result = dz_dt(1.0)
print(result)
```

```python
def z(t):
    x = jnp.e ** t
    y = jnp.log(t)
    return y*x**2 + x*y**3
dz_dt = grad(z, argnums=0)
result = dz_dt(1.0)
print(result)
print(jnp.e**2)
```
